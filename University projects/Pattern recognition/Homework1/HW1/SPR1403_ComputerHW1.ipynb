## Q1. Calculating Volume

```python
# Here you should implement volume of specified part of the image.
# input: radius of the sphere
# output: Area of the specified part
######################
#place your code here#
import math

def sphere_cube_volume(radius):
    # Calculate the volume of the sphere
    sphere_volume = (4/3) * math.pi * radius ** 3
    
    # Calculate the volume of the cube
    cube_volume = (2 * radius) ** 3
    
    # Calculate the volume of the area outside the sphere and inside the cube
    volume_difference = cube_volume - sphere_volume
    
    return round(volume_difference, 6)

def main():
    # Take the radius of the sphere as input
    radius = float(input("Enter the radius of the sphere: "))
    
    # Calculate and display the volume of the area outside the sphere and inside the cube
    result = sphere_cube_volume(radius)
    print("Volume of the area outside the sphere and inside the cube:", result)

if __name__ == "__main__":
    main()



######################
```

```python
import numpy as np
from scipy.stats import multivariate_normal
# Test

# Define mean vectors and covariance matrices for each class
mean_1 = np.array([-1, 1])
covariance_1 = np.array([[1, 0], [0, 1]])

mean_2 = np.array([1, 0])
covariance_2 = np.array([[2, 2], [2, 3]])

# Define prior probabilities
prior_1 = 0.5
prior_2 = 0.5

# Define a function to calculate class conditional probabilities
def class_conditional_probability(x, mean, covariance):
    return multivariate_normal.pdf(x, mean=mean, cov=covariance)

# Define a function to perform classification
def classify_sample(x):
    # Calculate class conditional probabilities
    p_x_given_1 = class_conditional_probability(x, mean_1, covariance_1)
    p_x_given_2 = class_conditional_probability(x, mean_2, covariance_2)
    
    # Calculate posterior probabilities using Bayes' theorem
    posterior_1 = p_x_given_1 * prior_1
    posterior_2 = p_x_given_2 * prior_2
    
    # Classify based on the class with the highest posterior probability
    if posterior_1 > posterior_2:
        return "Class ω1"
    else:
        return "Class ω2"

# Example usage:
sample = np.array([2, 3])  # Example sample to classify
classification = classify_sample(sample)
print("Sample", sample, "is classified as", classification)
```

```python
##### In this cell place answer of Q9.a.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Parameters
mu1 = np.array([8, 2])
mu2 = np.array([2, 8])
covariance = np.array([[4, 3], [3, 4]])
prior_prob = 0.5

# Define the Gaussian distributions for both classes
class1_distribution = multivariate_normal(mean=mu1, cov=covariance)
class2_distribution = multivariate_normal(mean=mu2, cov=covariance)

# Create a grid of points to evaluate the decision boundary
x = np.linspace(-5, 15, 400)
y = np.linspace(-5, 15, 400)
X, Y = np.meshgrid(x, y)
positions = np.vstack([X.ravel(), Y.ravel()])

# Calculate the PDF values for each class at each point
pdf_class1 = class1_distribution.pdf(positions.T).reshape(X.shape)
pdf_class2 = class2_distribution.pdf(positions.T).reshape(X.shape)

# Calculate the discriminant function
discriminant = np.log(pdf_class1) - np.log(pdf_class2)

# Plot the Bayesian decision boundary
plt.figure(figsize=(8, 6))
plt.contour(X, Y, discriminant, levels=[0], colors='black')
plt.scatter(mu1[0], mu1[1], color='blue', label='Class 1 Mean')
plt.scatter(mu2[0], mu2[1], color='red', label='Class 2 Mean')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bayesian Decision Boundary')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
